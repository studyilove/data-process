```python
af=df.groupby(['Outlet_Type','Item_Type'])['Item_Type'].agg(['count']).reset_index()
af
```

```python
af[af['Outlet_Type']=='Grocery Store'].sort_values(by='count',ascending=False)
```

```python
af[af['Outlet_Type']=='Supermarket Type1'].sort_values(by='count',ascending=False)
```

```python
df1=df.groupby(['Item_Type'])['Item_Type'].agg(['count'])
df1=df1.reset_index()
```

```python
df1.sort_values(by='count',ascending=False)
```

```python
plt.figure(figsize=(20,10))
x=df1['Item_Type']
y=df1['count']
plt.xlabel('Item_Type',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel('count',fontsize=14)
plt.bar(x,y)
plt.xticks(rotation=-35)
plt.show()
```

```python
dfa=df.groupby(['Item_Type'])['Item_MRP'].agg([('Item_MRP','sum')]).sort_values(by='Item_MRP',ascending=False).reset_index()
dfa
```

```python
plt.figure(figsize=(20,10))
x=dfa['Item_Type']
y=dfa['Item_MRP']
plt.xlabel('Item_Type',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel('Item_MRP',fontsize=14)
plt.bar(x,y)
plt.xticks(rotation=-35)
plt.show()
```

```python
df2=df.groupby(['Item_Fat_Content'])['Item_Fat_Content'].agg(['count'])
df2=df2.reset_index()
df2
```

```python

series = pd.Series(np.array([5517,3006]), index=['low fat','regular'],name='')
plt.title('Item_Fat_Content')
series.plot.pie(autopct='%.2f%%', fontsize=15,figsize=(6, 6))
```

```python
df3=df.groupby(['Item_Type'])['Item_Outlet_Sales'].agg(['sum']).reset_index()
df3
```

```python
plt.figure(figsize=(20,10))
x=df3['Item_Type']
y=df3['sum']
plt.xlabel('Item_Type',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel('sales',fontsize=14)
plt.bar(x,y)
plt.xticks(rotation=-35)
plt.show()
```
